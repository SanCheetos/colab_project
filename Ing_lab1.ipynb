```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis",
 "blanchefort/rubert-base-cased-sentiment")
classifier("Я обожаю инженерию машинного обучения!")
```

```python
classifier("Мне, наверное, нравится инженерия машинного обучения!")
```

```python
classifierMed = pipeline("sentiment-analysis",
 "blanchefort/rubert-base-cased-sentiment-med")
classifierMed("У вас рак")
```

```python
classifierMed("Вы полностью здоровы")
```

```python
classifierRuS = pipeline("sentiment-analysis",
 "blanchefort/rubert-base-cased-sentiment-rusentiment")
classifierRuS("Я русский")
```

```python
classifierRuS("Я сегодня проснулся вялым")
```
